```python
# Import a bunch of libraries
import pandas as pd
import numpy as np
import statsmodels as sm
import plotly.express as px
import plotly.figure_factory as ff
from scipy import stats
```

```python
# generate the 1000 observations from a bunomial distribution
obs = np.random.binomial(30, 0.2, size=1000)
df = pd.DataFrame(data=obs, columns=["ybinom"])
df.describe()
```

```python
# The mean looks about right so now lets use a histogram to visualize the data
hist = px.histogram(df, x="ybinom")
hist.show()
```
